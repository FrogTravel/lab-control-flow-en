## Words

```python
words = ['play', 'filling', 'bar', 'theatre', 'easygoing', 'date', 'lead', 'that', 'story',  'island']
```

**Print every word in upper case**

```python
upper_words = [word.upper() for word in words]
print(upper_words)
```

**Create a new list containing only words with 5 or more letters**

```python
long_words = [word for word in words if len(word) >= 5]
print(long_words)
```

**Print the first word starting with "t"**

```python
for word in words:
    if word.startswith("t"):
        print(word)
        break
```

## Numbers

**Create a list containing the square of every number from 1 to 10**

```python
squares = [number**2 for number in range(1, 11)]
print(squares)
```

**Print a list containing the square of every odd number from 1 to 10**

```python
squares_of_odd = [number**2 for number in range(1, 11, 2)]
print(squares_of_odd)
```

**Create a list with the squares of all multiples of 8 below 1000**

```python
eight_mult_square = [number**2 for number in range(1000, 0, -1) if number % 8 == 0]
print(eight_mult_square)
```

## People

```python
people = [
    {
        "name": "Juan",
        "age": 34,
        "n_kids": 2
    },
    {
        "name": "Pepe",
        "age": 27,
        "n_kids": 0
    },
    {
        "name": "Sonia",
        "age": 41,
        "n_kids": 1
    },
    {
        "name": "Lucía",
        "age": 22,
        "n_kids": 2
    },
    {
        "name": "Leo",
        "age": 55,
        "n_kids": 5
    }
]
```

**How many people are there?**

```python
print(len(people))
```

**How many people have kids**?

```python
parents = len([x["n_kids"] for x in people if x["n_kids"] > 0])
print(parents)
```

**How many kids do they have in total?**

```python
total_number_of_kids = sum([x["n_kids"] for x in people])
print(total_number_of_kids)
```

**In a year's time, names ending with "a" will have an extra kid. Create a list of dictionaries with people's info in a year's time**

```python
a_people = [item for item in people if item["name"][-1] == "a"]
new_people = people.copy()
for human in new_people:
    if human in a_people:
        human.update({"n_kids" : human["n_kids"] + 1})
new_people
```
